## 2. Excluding some features

## 3. Identify columns that have over 85% missing values

>I am going to create a loop to iterate through all the columns of the categorical data and append the columns that have 85% or more of empty values to the drop_list we created before.

```python
total_columns_dropped = 0  # I put a count to store the number of columns that have more than 85% of empty spaces
for column in categorical:
    categorical[column] = categorical[column].apply(str) # Changing any possible integer or float value to string
    space_count = 0  # This count stores the total nomber of empty spaces per feature
    for value in categorical[column]:
        x = value.isspace()  # Checking if all the string is a space (e.g "   e   " would be False)
        if x == True:
            space_count += 1
    if space_count/len(categorical[column]) >= 0.85:  # Checking the proportion of spaces there are in the feature
        drop_list.append(column)
        total_columns_dropped += 1
    space_count = 0  # reseting the count
print("")
print('\033[1m'"There are {} columns with more than 85% of empty values"'\033[0m'.format(total_columns_dropped))
print("")
```

>Building the loop, I observed that some of the columns had integers or floats as values even tho the column type was an object and it was giving me troubles (<span style="color: red">AttributeError:</span>  'int' object has no attribute 'isspace') that is why I decided to include code row to transform all the values of those columns into string to avoid it.

```python
drop_list # Sanity check
```

## 4. Remove those columns from the dataframe

```python
categorical = categorical.drop(columns=drop_list, axis=0)
```

```python
categorical
```

## 5. Reduce the number of categories in the column GENDER

```python
# We will replace the null values of column 'GENDER' with 'F' as it was said in the Instructions

categorical['GENDER'].value_counts()
```

```python
categorical['GENDER'] = categorical['GENDER'].fillna('F')
```

```python
categorical['GENDER'].value_counts() #Sanity check
```

**This does not produce anything because there are no nans but empty spaces**

```python
# Replacing empty spaces for 'F' and the rest for 'other' and keeping 'M'

categorical['GENDER'] = categorical['GENDER'].replace(' ', 'F').replace(['U', 'J', 'C', 'A'], "other")
```
